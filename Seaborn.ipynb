## 1: Write a Python program to create a plot to get a general Statistics of Iris data. (Using the dataset “iris.csv”).

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import preprocessing 
iris = pd.read_csv("iris.csv")
iris
```

```python
iris.describe().plot(kind='area')
plt.xlabel("Statistics")
plt.ylabel("Value")
```

## 2: Write a Python program to create a graph to find relationship between the sepal length and width. (Using the dataset “iris.csv”).

```python
# sns.scatterplot(data=iris, x='sepal_length', y='petal_width', hue='species')
df=sns.load_dataset("iris")
sns.scatterplot(x='sepal_length',y='sepal_width',data=df)
plt.show()
```

## 3: Write a Python program to create a join plot using "hexbin" to describe individual distributions on the same plot between Sepal length and Sepal width. (Using the dataset “iris.csv”).

```python
# sns.jointplot(x="sepal.length",y="sepal.width",data=iris,kind='hex',color='orange')
sns.jointplot(x='sepal_length',y='sepal_width',data=df,kind='hex')
plt.show()
```

## 4: Write a Python program to create a join plot and add regression and kernel density fits using "reg" to describe individual distributions on the same plot between Sepal length and Sepal width. (Using the dataset “iris.csv”).

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=df,kind='reg')
plt.show()
```
