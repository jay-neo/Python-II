## 1: Take a data set in a data frame and plot the bar chart example may be year of Production and the Production Quantity. (Take help of Pandas)

```python
import pandas as pd
import matplotlib.pyplot as plt
data={
'Year of Production':[2018,2019,2020,2021,2022],
'Production Quantity':[250,190,100,120,210]
}
df=pd.DataFrame(data)
df.plot(x="Year of Production",y="Production Quantity",kind='bar',legend=False)
plt.xlabel("Year of Production")
plt.ylabel("Production Quantity")
plt.title("Production Quantity vs Year of Production ")
plt.show()
```

## 2: Take a data set from the above data frame which you have prepared and plot the pie chart. (Take help of Pandas)

```python
import pandas as pd
import matplotlib.pyplot as plt
data={
'Year of Production':[2018,2019,2020,2021,2022],
'Production Quantity':[250,190,100,120,210]
}
df=pd.DataFrame(data)
df.plot(x="Year of Production",y="Production Quantity",kind='pie',legend=False,autopct='%1.1f%%')
plt.xlabel("Year of Production")
plt.ylabel("Production Quantity")
plt.title("Production Quantity vs Year of Production ")
plt.show()
```

## 3: Take a csv file with name, employee id and salary. Read the data of salary and plot the bar Chart with employee name on the x axis and employee salary on the y-axis. Just take 5 records in the csv files.

```python
import pandas as pd
data={
'Name':['Alexa','Rohit','Shubhomay','Shubhodeep','Aritra'],
'Employee Id':[100,101,102,103,104],
'Salary':[200000,30000,400000,50000,80000]
}
df=pd.DataFrame(data)
df.to_csv("employee.csv",index=False)
import matplotlib.pyplot as plt
df=pd.read_csv("employee.csv")
plt.figure(figsize=(10,6))
plt.bar(df["Name"],df["Salary"])
plt.xlabel("Employee Name")
plt.ylabel("Salary")
plt.title("Employee name with bar chart")
plt.show()
```

## 6: Take x values and plot the cos(x) and tan(x) in a graph using matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,x,z)
plt.xlabel('x values from 0 to 4 pi')
plt.ylabel('sin(x) and cos(x)')
plt.title('Plot of sin and cos from 0 to 4pi')
plt.legend(['sin(x)', 'cos(x)'])
plt.show()
```
